```python
from ultralytics import YOLO
import os

# Load model
model = YOLO('yolov8n.pt')

# Folder containing images
input_folder = "/content/drive/MyDrive/INPUTS"

# Output folder
output_folder = "/content/drive/MyDrive/OUTPUTS"
os.makedirs(output_folder, exist_ok=True)

# Process all images
for filename in os.listdir(input_folder):

    if filename.endswith((".jpg", ".png", ".jpeg")):

        image_path = os.path.join(input_folder, filename)

        # Run detection
        results = model(image_path)

        # Save result
        save_path = os.path.join(output_folder, filename)
        results[0].save(filename=save_path)

        print(f"Done: {filename}")

print("All images processed!")
```

```python
from ultralytics import YOLO
import cv2
import matplotlib.pyplot as plt

model = YOLO('/content/drive/MyDrive/YOLO_MODEL/best.pt')

results = model('/content/drive/MyDrive/INPUTS/14728.jpg', conf=0.92,iou=0.45)

img = cv2.imread('/content/drive/MyDrive/INPUTS/14728.jpg')
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

red_img = img.copy()
red_img[:, :, 0] = img[:, :, 2]
red_img[:, :, 2] = img[:, :, 0]

boxes = results[0].boxes.xyxy.cpu().numpy()

for box in boxes:
    x1, y1, x2, y2 = map(int, box)

    cv2.rectangle(
        red_img,
        (x1, y1),
        (x2, y2),
        (0, 0, 255),
        2
    )

plt.figure(figsize=(12,8))
plt.imshow(red_img)
plt.axis('off')
plt.title("YOLOv8 Detection")
plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

img = cv2.imread('/content/drive/MyDrive/INPUTS/14728.jpg')
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

hsv = cv2.cvtColor(img, cv2.COLOR_RGB2HSV)

lower_blue = np.array([100, 80, 80])
upper_blue = np.array([140, 255, 255])

mask = cv2.inRange(hsv, lower_blue, upper_blue)

kernel = np.ones((3,3), np.uint8)
mask = cv2.morphologyEx(mask, cv2.MORPH_OPEN, kernel)

contours, _ = cv2.findContours(
    mask,
    cv2.RETR_EXTERNAL,
    cv2.CHAIN_APPROX_SIMPLE
)

output = img.copy()

for cnt in contours:
    area = cv2.contourArea(cnt)

    if area > 80:
        x, y, w, h = cv2.boundingRect(cnt)

        cv2.rectangle(
            output,
            (x, y),
            (x + w, y + h),
            (255, 0, 0),
            2
        )

plt.figure(figsize=(12,8))
plt.imshow(output)
plt.title("BDA Cell Detection")
plt.axis('off')
plt.show()
```

```python
import time
import cv2
import numpy as np
import matplotlib.pyplot as plt
import torch
import celldetection as cd
from ultralytics import YOLO

image_path = '/content/drive/MyDrive/INPUTS/14728.jpg'

model_yolo = YOLO('/content/drive/MyDrive/YOLO_MODEL/best.pt')

device = 'cuda' if torch.cuda.is_available() else 'cpu'

model_bda = cd.fetch_model(
    'ginoro_CpnResNeXt101UNet-fbe875f1a3e5ce2c',
    check_hash=True
).to(device)

model_bda.eval()

start = time.time()

results = model_yolo(
    image_path,
    conf=0.92,
    iou=0.45
)

yolo_time = time.time() - start

img = cv2.imread(image_path)
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

yolo_img = img.copy()

yolo_img[:, :, 0] = img[:, :, 2]
yolo_img[:, :, 2] = img[:, :, 0]

boxes = results[0].boxes.xyxy.cpu().numpy()

for box in boxes:
    x1, y1, x2, y2 = map(int, box)

    cv2.rectangle(
        yolo_img,
        (x1, y1),
        (x2, y2),
        (0, 0, 255),
        2
    )

start = time.time()

with torch.no_grad():

    x = cd.to_tensor(
        img,
        transpose=True,
        device=device,
        dtype=torch.float32
    )

    x = x / 255
    x = x[None]

    y = model_bda(x)

bda_time = time.time() - start

bda_img = img.copy()

hsv = cv2.cvtColor(img, cv2.COLOR_RGB2HSV)

lower_blue = np.array([100, 80, 80])
upper_blue = np.array([140, 255, 255])

mask = cv2.inRange(hsv, lower_blue, upper_blue)

contours, _ = cv2.findContours(
    mask,
    cv2.RETR_EXTERNAL,
    cv2.CHAIN_APPROX_SIMPLE
)

for cnt in contours:

    area = cv2.contourArea(cnt)

    if area > 100:

        x, y, w, h = cv2.boundingRect(cnt)

        cv2.rectangle(
            bda_img,
            (x, y),
            (x + w, y + h),
            (255, 0, 0),
            2
        )

fig, axes = plt.subplots(1, 2, figsize=(16, 7))

axes[0].imshow(yolo_img)
axes[0].set_title(f'YOLOv8 Detection\nTime: {yolo_time:.2f}s')
axes[0].axis('off')

axes[1].imshow(bda_img)
axes[1].set_title(f'BDA Detection\nTime: {bda_time:.2f}s')
axes[1].axis('off')

plt.suptitle(
    'Comparative Analysis - Brain Cell Detection',
    fontsize=16
)

plt.tight_layout()

plt.savefig('comparison_output.jpg')

plt.show()
```
